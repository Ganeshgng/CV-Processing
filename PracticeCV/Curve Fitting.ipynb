```python
import numpy as np
import matplotlib.pyplot as plt

# defining actual values
m = 2
c = 1

x = np.arange(0,100,0.5)
line = m*x +c
points = line + np.random.normal(0,100,len(x))

# Curve fitting
Y = points.reshape(len(points), 1)
X = np.concatenate((x.reshape((len(x),1)), np.ones((len(x),1))), axis=1)
coefficients = np.linalg.pinv(X.T@X)@X.T@Y
Y_ = X@coefficients

plt.plot(x, line, color = 'r')
plt.scatter(x, points)
plt.plot(x, Y_, color = 'g')
plt.show()
```
